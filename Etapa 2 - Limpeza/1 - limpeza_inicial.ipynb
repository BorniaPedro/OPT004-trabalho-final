```python
import pandas as pd

df = pd.read_csv('./Crimes_2019_to_2024.csv')
```

```python
# onde 'Location Description' é nulo, completa com 'UNSPECIFIED'
df.fillna({'Location Description': 'UNSPECIFIED'}, inplace=True)

# apaga linhas aonde as colunas 'Ward' ou 'Community Area' são nulas
df.dropna(subset=['Ward', 'Community Area'], inplace=True)
```

```python
# Remove colunas que nao vou ser utilizadas
colunas_para_remover = [
    'Case Number', 'Block', 'IUCR', 'Beat', 'District', 'Ward', 
    'FBI Code', 'Updated On', 'X Coordinate', 'Y Coordinate', 
    'Latitude', 'Longitude', 'Location'
]

df.drop(columns=colunas_para_remover, inplace=True, errors='ignore')
```

```python
# formata a data
df['datetime'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p')

# cria coluna so com data, pra integração
df['date_key'] = df['datetime'].dt.normalize()

# garante que community area é inteiro
df['Community Area'] = df['Community Area'].astype(int)
```

```python
# exporta em csv o dataframe limpo
df.to_csv('./Crimes_2019_to_2024_Limpo.csv', index=False)
```
