### Seleciona apenas os dados de 2019 a 2024
download em "https://catalog.data.gov/dataset/crimes-2001-to-present"

```python
import pandas as pd
import numpy as np

df = pd.read_csv('./Crimes_-_2001_to_Present.csv')

# Cria um dataset selecionando apenas os crimes entre 2019 e 2024
df_2019_2024 = df[(df['Year'] >= 2019) & (df['Year'] <= 2024)]

# Exporta pra um novo arquivo CSV
df_2019_2024.to_csv('./Crimes_2019_to_2024.csv', index=False)
```
